````python
from scipy import misc
from scipy import ndimage
import cv2
import numpy as np
import matplotlib.pyplot as plt
````

````python
img = cv2.imread('imagens/baboon.png')
````

````python
img.shape
````

## a)

Aqui é bem autoexplicativo. É criado uma matriz que irá multiplicar os vetores que representam os três canais de cores de cada pixel.

````python
matriz_a = np.array([[0.393, 0.769, 0.189],
                     [0.394, 0.686, 0.168],
                     [0.272, 0.534, 0.131]])
````

````python
img_a = np.dot(img, matriz_a)/255
img_a = img_a.clip(max=[1,1,1])
img_a.shape
````

````python
plt.imshow(img_a)
````

## b)

Semelhante ao item "a", porém agora faremos uma multiplicação vetorial que nos resultará em uma imagem de canal único (imagem cinza).

````python
vetor_b = np.array([0.2989, 0.5870, 0.1140])
````

````python
img_b = np.tensordot(img, vetor_b, axes=([2], [0]))/255
img_b = img_b.clip(max=[1]).reshape(img.shape[0:2])
img_b.shape
````

````python
plt.imshow(img_b)
````
